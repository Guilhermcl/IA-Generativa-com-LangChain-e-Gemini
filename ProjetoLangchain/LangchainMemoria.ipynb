```python
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_core.runnables.history import RunnableWithMessageHistory
from langchain_core.chat_history import InMemoryChatMessageHistory
from langchain_google_genai import ChatGoogleGenerativeAI 
# template base para conversação

prompt = ChatPromptTemplate.from_messages([
    ("system", "Você é um assistente útil. Responda de forma natural e direta."),
    MessagesPlaceholder(variable_name="chat_history"),
    ("human", "{input}")
])

chain = prompt | gemini


def test_conversation(conversation, memory_store, memory_type):
    """Função de teste"""
    print(f"\n{'='*60}")
    print(f"TESTANDO: {memory_type}")
    print(f"{'='*60}")
   
    questions = [
        "Meu nome é Maria e sou engenheira de software.",
        "Trabalho numa empresa de tecnologia em São Paulo.",
        "Tenho 28 anos e adoro programar em Python.",
        "Qual é minha profissão?",
        "Onde eu trabalho?",
        "Quantos anos eu tenho?",S
        "Qual linguagem eu gosto de programar?"
    ]
   
    session_id = "test_session"
   
    for i, question in enumerate(questions, 1):
        response = conversation.invoke(
            {"input": question},
            config={"configurable": {"session_id": session_id}}
        )
        print(f"\n{i}. Usuário: {question}")
        print(f"   Gemini: {response.content}")
```

```python
#MEMÓRIA BUFFER COMPLETA (guarda tudo)
print("1. ConversationBufferMemory - Guarda toda a conversa")

# Armazenamento que guarda TUDO
memory_buffer_store = {}

def get_buffer_history(session_id: str):
    """Guarda toda a conversa sem limite"""
    if session_id not in memory_buffer_store:
        memory_buffer_store[session_id] = InMemoryChatMessageHistory()
    return memory_buffer_store[session_id]

conversation_buffer = RunnableWithMessageHistory(
    chain,
    get_buffer_history,
    input_messages_key="input",
    history_messages_key="chat_history"
)

test_conversation(conversation_buffer, memory_buffer_store, "MEMÓRIA BUFFER COMPLETA")
```
